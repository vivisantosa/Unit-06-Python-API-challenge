### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
vacation_df = cities_df
vacation_df = vacation_df[(vacation_df["Max Temp"] >= 28) & (vacation_df["Max Temp"] <= 35)]
vacation_df = vacation_df[(vacation_df["Wind Speed"] > 1)& (vacation_df["Wind Speed"] < 5)]
vacation_df = vacation_df[(vacation_df["Cloudiness"] < 20)]
vacation_df 
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel1_df = vacation_df
hotel1_df['Hotel Name'] = ""
hotel1_df.reset_index(drop=True, inplace=True)
hotel1_df
```

```python
# find the closest hotel of each city

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = { 
    "radius": 5000,
    "type": "hotel",
    "keyword":"bed-breakfast",
    "key": g_key               }

for i in range(0,len(hotel1_df)):    
    city = hotel1_df.iloc[i,0]
    lat = hotel1_df.iloc[i,5]                  #get location
    lng = hotel1_df.iloc[i,6]
    params['location'] = f"{lat},{lng}"       # add location to params dict
    
    response = requests.get(base_url, params=params).json()   # assemble url and make API request
    results = response['results']                             # extract results (results is a dict in googleapi)
    #print(response)
    
    if response['status'] == 'OK':                           # Result is not []     
        print(f"Hotel in {city} is {results[0]['name']}.")
        hotel1_df.iloc[i,9] = results[0]['name']          # add to the column
    
    else :
        print(f"NO HOTEL IN {city}.")
```

```python
hotel1_df
```

```python
hotel_df = hotel1_df[hotel1_df['Hotel Name'] != '']
hotel_df = hotel_df.drop(["Cloudiness","Date","Humidity","Max Temp","Wind Speed"], axis=1)
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations,info_box_content=hotel_info)

# Display figure
fig.add_layer(markers)
fig
```
